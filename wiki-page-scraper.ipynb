#### This ipynb notebook helps the user to fetch all the official homepages/websites of the links mentioned in the current wiki page which normally on clicking opens up a wiki page. But my code will help the user to list all the official homepage links.

```python
#import the library to query the website and get the html page of the webpage
from urllib.request import urlopen

#here give the link of the site you want to scrape all external links
page = urlopen("https://en.wikipedia.org/wiki/List_of_biological_databases")
```

```python
#beautifulsoup library to parse the html page
from bs4 import BeautifulSoup
soup = BeautifulSoup(page)
```

```python
#create empty lists:
official_website = {}    
```

```python
#here for each link in the mentioned page, open that link-page and get its official website link
for link in soup.find_all("a"):
    if(link.get("href")):
        if("wiki" in link.get("href")):
            try:
                interpage = urlopen("https://en.wikipedia.org"+link.get("href"))
                interpage_Sp = BeautifulSoup(interpage)
                for link2 in interpage_Sp.find_all("a"):
                    if((("Official website" in link2.get_text()) or ("home page" in link2.get_text())) and (''.join(link2.get("class"))=="externaltext") and (link2.get("rel")[0]=="nofollow")):
                        official_website[link2.get('href')] = link.get_text()
                        print(official_website[link2.get('href')])
                        print("done")
            except:
                pass
```

```python
#create table
import pandas as pd
df = pd.DataFrame()
df['Name'] = official_website.values()
df['Link'] = official_website.keys()
```

```python
#save it to excel
writer = pd.ExcelWriter('BI_Tables.xlsx')
df.to_excel(writer,'sheet1')
writer.save()
```
